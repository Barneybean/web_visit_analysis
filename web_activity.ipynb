```python
del file[0:7]
file
```

```python
length = len(file)

# initialize count for users visit 12 and 17 in same session
count_both = 0

for line in file:
    if re.search( r'12', line) and re.search( r'17', line):
        count_both += 1 

Perc_both_page = count_both/length

print("{0:.2%}".format(Perc_both_page))
```

```python
# initialize count for users visit 12 and 17 in same session
count_sequence = 0

for line in file:
    if re.search( r'17.*12', line):
        count_sequence += 1 

Perc_sequence = count_sequence/length

print("{0:.2%}".format(Perc_sequence))
```
